```python
# Import basics for qiskit 
from qiskit import QuantumCircuit, transpile
from qiskit.circuit import QuantumCircuit
from qiskit.converters import circuit_to_dag, dag_to_circuit
```

```python
# Create circuit
qc = QuantumCircuit(3, 2)

# Make teleportation circuit
qc.h(1)
qc.cx(1, 2)

qc.cx(0, 1)
qc.h(0)
qc.x(2)

# Measure first two qubits
qc.measure(0, 0)
qc.measure(1, 1)

# Barrier
qc.barrier()

# Apply classically controlled gates
with qc.if_test((1,1)):
    qc.x(2)
    qc.x(1)
    with qc.if_test((1,1)):
        qc.x(2)
        qc.x(1)

# Draw circuit
import matplotlib.pyplot as plt
qc.draw('mpl')
```

```python
# Transpile circuit
backend = 'qasm_simulator'

# Transpile the circuit for the backend
compiled_circuit = transpile(qc, optimization_level=3)

# Draw the transpiled circuit
compiled_circuit.draw('mpl')
```

```python
# Convert circuit to DAG
from qiskit import QuantumCircuit
from qiskit.converters import circuit_to_dag
dag = circuit_to_dag(qc)

# Draw the DAG
dag.draw()
```
